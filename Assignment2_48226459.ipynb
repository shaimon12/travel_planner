## TinyLLaMA Integration (Local LLM )

To complement the API-based Gemini setup, we also demonstrate a **local LLM** using Hugging Face’s `transformers` library and a lightweight LLaMA-compatible model.

**Model Used:** TinyLLaMA 1.1B Chat  
**Why:** It's resource-friendly, fast to load.

### Key Features:
- Runs entirely offline (no internet or API required)
- Utilizes GPU if available, or falls back to CPU
- Handles basic travel prompts and generates coherent itineraries

This setup shows how LLaMA-style models can serve as a **privacy-preserving, cost-free alternative** to cloud-based LLMs, making them a strong candidate for hybrid assistant systems.


````python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

tokenizer = AutoTokenizer.from_pretrained(model_id, use_fast=False)
tinyllama_model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    device_map="auto"
)

prompt = "Plan a 2-day trip to Paris for someone interested in art, history, and food."
inputs = tokenizer(prompt, return_tensors="pt").to(tinyllama_model.device)
output = tinyllama_model.generate(**inputs, max_new_tokens=150)

print(tokenizer.decode(output[0], skip_special_tokens=True))
````

### TinyLLaMA Output: Paris Itinerary Analysis

To complement Gemini, we tested a lightweight, offline model — **TinyLLaMA 1.1B Chat** — using Hugging Face’s `transformers`. It runs locally with no internet or API needed, making it ideal for privacy-focused or low-resource environments.

**Model:** `TinyLlama/TinyLlama-1.1B-Chat-v1.0`  
**Prompt:** _"Plan a 2-day trip to Paris for someone interested in art, history, and food."_

#### Response:
2. **Paris: A Cultural and Food Tour** – This tour covers the best of Paris, including the Eiffel Tower, Notre-Dame Cathedral, and the Louvre Museum.  
3. **Paris: A Food Tour** – This tour takes you to some of the best restaurants in Paris, including brunch at Le Jules Verne and dinner at Le Comptoir du Relais.  
4. **Paris: A Food and History Tour** – This tour combines both food and history, with visits to famous Parisian restaurants and museums.  
5. **Paris: A Food and Culture Tour** – This tour focuses on the best of Parisian cuisine and culture, with visits to famous...

> **Note:** The response begins at “2.” due to the model interpreting the prompt as part of a broader list.

Despite its simplicity, TinyLLaMA produced relevant suggestions with real locations and food spots. While not as polished as Gemini, it’s a solid choice for basic offline itinerary generation.


````python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load TinyLLaMA model
model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
tokenizer = AutoTokenizer.from_pretrained(model_id, use_fast=False)
tinyllama_model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    device_map="auto"
)

# Helper function to generate and print output
def generate_and_print(prompt_text, label):
    inputs = tokenizer(prompt_text, return_tensors="pt").to(tinyllama_model.device)
    output = tinyllama_model.generate(**inputs, max_new_tokens=300, temperature=0.8, top_p=0.9, do_sample=True)
    result = tokenizer.decode(output[0], skip_special_tokens=True)
    print(f"=== Prompt {label} ===\n")
    print(result)
    print("\n" + "="*80 + "\n")

# -----------------------------
# PROMPT 1: Kyoto Temples & Food
# -----------------------------
prompt_1 = "Plan a 3-day trip to Kyoto for someone interested in temples and food."
generate_and_print(prompt_1, "1: Kyoto")

# -----------------------------
# PROMPT 2: Family Beach Vacation
# -----------------------------
prompt_2 = '''You are a helpful travel assistant.

Classify the following travel request and generate a 4-day itinerary accordingly:
"I want a beach vacation in Southeast Asia with activities for my two kids, ages 5 and 9."'''
generate_and_print(prompt_2, "2: Southeast Asia Family Beach Trip")

# -----------------------------
# PROMPT 3: Adventure in New Zealand
# -----------------------------
prompt_3 = '''Instruction: You are a travel expert assistant. Generate a personalized adventure plan.

User: I want an adrenaline-filled trip to New Zealand. Include bungee jumping, hiking, and rafting.

Assistant:'''
generate_and_print(prompt_3, "3: New Zealand Adventure")
````

### Prompt Response Analysis (TinyLLaMA)

TinyLLaMA handled all three prompts reasonably well, considering its smaller size and local, resource-efficient nature. While responses were less structured and sometimes incomplete, the model showed an ability to interpret core themes and return relevant suggestions.

# Assistant Dialogue System

This section presents an interactive travel assistant that simulates a dialogue with users to collect personal preferences, travel goals, and specific constraints. Based on the inputs, it generates personalized travel itineraries using both public (Gemini API) and local (TinyLLaMA) LLMs.


## Task Definition – Introducing the Travel Assistant

The first step in the dialogue is to help the user understand the role of the assistant.

The assistant opens the conversation by introducing itself and explaining what kind of help it can provide. This ensures that the user clearly understands the assistant’s capabilities and can begin offering relevant details.

This aligns with the requirement to demonstrate the assistant’s task definition in a conversational context.


````python
# Assistant opening message – explaining its purpose
def get_task_definition_message():
    return (
        "Hello! I'm your personal travel assistant. "
        "I can help you plan a trip by learning about your preferences, budget, destination, and more. "
        "I'll ask you a few quick questions, and then I’ll generate a customized travel itinerary just for you. "
        "Ready to start?"
    )

# Simulate assistant speaking
intro_message = get_task_definition_message()
print("Assistant:", intro_message)
````

## Analysis: Task Definition

The assistant introduces itself clearly and sets the context for the dialogue:

> "I'm your personal travel assistant... I’ll ask you a few quick questions, and then I’ll generate a customized travel itinerary..."

This covers the required **task definition** by explaining what the assistant does and what the user can expect.


## Collecting Personal Information

The assistant now asks for the user’s basic personal details to help personalize the itinerary. This step collects inputs like name, age, and travel style.

This simulates the beginning of the conversation where the assistant gathers information to tailor its recommendations.


````python
# Simulate dialogue input (in a real app, this would come from user interaction)
def collect_personal_info():
    print("Assistant: Great! Let’s start with some quick details to personalize your trip.")
    name = "Ali"  # Simulated user input
    age = 28       # Simulated user input
    travel_type = "solo"  # Simulated user input

    print(f"User: My name is {name}, I’m {age} years old, and I’ll be traveling {travel_type}.")

    return {
        "name": name,
        "age": age,
        "travel_type": travel_type
    }

# Run the function to simulate the assistant asking and the user replying
personal_info = collect_personal_info()
````

## Analysis: Personal Information

The assistant gathers basic personal information, including name, age, and travel type:

> "My name is Ali, I’m 28 years old, and I’ll be traveling solo."

This input allows the assistant to generate more relevant and personalized travel recommendations. The approach reflects practical work, where dialogue models collected structured input before generating a response.


## Collecting Travel Interests

Next, the assistant asks about the user’s travel interests. This helps tailor the itinerary to match the kind of activities the user enjoys, such as sightseeing, food, history, adventure, or relaxation.


````python
# Continue the simulated dialogue to collect travel interests
def collect_travel_interests():
    print("Assistant: Awesome! What kind of activities or experiences are you most interested in during your trip?")
    interests = ["museums", "local food", "architecture"]  # Simulated user input
    print(f"User: I’m mostly interested in {', '.join(interests)}.")
    return interests

# Run and store user interests
user_interests = collect_travel_interests()
````

## Analysis: Travel Interests

The assistant asks about the user’s interests:

> "I’m mostly interested in museums, local food, architecture."

This helps guide the LLM to focus on relevant activities in the final itinerary. The structure follows the conversational format, where user input is collected step by step to inform LLM responses.


## Collecting Additional Preferences

The assistant now gathers more specific details such as the destination, budget, and trip duration. These inputs help shape the structure and scope of the final travel plan.


````python
# Continue simulated dialogue for destination, budget, and duration
def collect_additional_preferences():
    print("Assistant: Great! Where are you planning to travel, what’s your budget, and how many days will you be staying?")
    destination = "Barcelona"
    budget = "moderate"
    duration_days = 3
    print(f"User: I'm going to {destination}, my budget is {budget}, and I’ll stay for {duration_days} days.")
    return {
        "destination": destination,
        "budget": budget,
        "duration_days": duration_days
    }

# Run and store preferences
additional_prefs = collect_additional_preferences()
````

## Analysis: Additional Preferences

The assistant gathers key travel specifics:

> "I'm going to Barcelona, my budget is moderate, and I’ll stay for 3 days."

This gives the assistant enough context to generate a customized itinerary. It completes the dialogue chain by combining user identity, interests, and logistics — exactly what’s needed before passing the data to an LLM.


## Itinerary Generation with LLM

Now that the assistant has collected all the necessary information, it uses a prompt template to send the data to the LLM (Gemini and TinyLlama) and generate a personalized 3-day itinerary for the user.


````python
# Combine all user inputs into one dictionary
user_data = {
    **personal_info,
    "interests": user_interests,
    **additional_prefs
}

# Prompt builder based on collected inputs
def build_prompt(user):
    return f"""
You are a travel assistant. Create a detailed {user['duration_days']}-day itinerary for a {user['travel_type']} traveler named {user['name']}, aged {user['age']}, visiting {user['destination']}.

They are interested in {", ".join(user['interests'])}, with a {user['budget']} budget.

Provide morning, afternoon, and evening activities for each day. Include food suggestions and local cultural highlights.
"""

# Generate prompt and send to Gemini
prompt = build_prompt(user_data)
print("Prompt Sent to Gemini:\n", prompt)

# Get Gemini response
gemini_response = gemini_model.generate_content(prompt)
print("\nGemini's Travel Plan:\n")
print(gemini_response.text)
````

## Itinerary Generation: Gemini (Cloud-Based LLM)

Gemini successfully generated a highly detailed 3-day itinerary for a solo traveler visiting Barcelona, incorporating the user's interests (**museums, food, architecture**), **budget level (moderate)**, and **trip length (3 days)**.

### The response includes:
- Morning, afternoon, and evening activity breakdowns  
- Named food recommendations and local cultural experiences  
- Transportation advice and safety tips (e.g., metro passes, pickpocket warnings)  
- Context-aware recommendations like booking in advance and dress codes  

### Analysis:
The format is **clean**, **rich in local insights**, and **matches user intent effectively**. Gemini's ability to adapt tone, include **practical travel tips**, and structure output logically demonstrates strong **prompt-following and planning capabilities**. It is highly suited for **production-grade applications** where detailed, polished content is expected.


````python
# Reuse the same prompt from Step 5
llama_prompt = build_prompt(user_data)

# Tokenize and generate using TinyLLaMA
inputs = tokenizer(llama_prompt, return_tensors="pt").to(tinyllama_model.device)
output = tinyllama_model.generate(**inputs, max_new_tokens=500)

# Decode and display result
llama_output = tokenizer.decode(output[0], skip_special_tokens=True)
print("TinyLLaMA's Travel Plan:\n")
print(llama_output)
````

## Itinerary Generation: TinyLLaMA (Local LLM)

TinyLLaMA produced a complete 3-day itinerary for a solo traveler visiting Barcelona based on user inputs (name, age, destination, interests, and budget). It maintained consistent structure and logical activity flow throughout.

> **Note:** The maximum token limit was set to 500 to accommodate TinyLLaMA’s constraints with longer response generation. This influenced the brevity and repetition in the output.

### The response includes:
- Structured daily itineraries with **Morning, Afternoon, and Evening** segments  
- Relevant activities and popular locations (e.g., **Sagrada Familia**, **Park Güell**, **Fundació Joan Miró**)  
- General food experiences like **tapas bars** and **local cafés**  
- Repeated walking suggestions (e.g., along **Barceloneta beach**, **Eixample district**, **Passeig de Gracia**)  

### Analysis:
While simpler in language and somewhat repetitive in phrasing, the itinerary aligns well with the prompt and includes legitimate points of interest. TinyLLaMA does not provide specific restaurant names, cultural tips, or transportation guidance, which limits its depth and personalization. However, it performs well as a **resource-light, fully offline model** capable of supporting **basic travel planning**.

This example demonstrates how local LLMs like TinyLLaMA can generate **coherent, useful travel itineraries** without relying on internet connectivity or cloud resources, even under **token limitations**.


## Comparison: Gemini vs. TinyLLaMA (Notebook)

To evaluate the strengths and limitations of different LLMs in generating travel itineraries, we compared the outputs of a cloud-based model (**Gemini 1.5 Pro**) and a local model (**TinyLLaMA 1.1B Chat**) using the same prompt and user profile.

> **Note:** TinyLLaMA was configured with a maximum token limit of 500 to ensure a concise response and avoid generation issues common with longer outputs in smaller models.

| Feature                     | Gemini 1.5 Pro (Cloud/API)                          | TinyLLaMA 1.1B (Local)                           |
|-----------------------------|-----------------------------------------------------|--------------------------------------------------|
| **Access**                  | Requires internet and Google API key               | Fully offline via Hugging Face Transformers      |
| **Response Detail**         | Rich, structured, and context-aware                | Basic, clear structure but less nuanced          |
| **Named Recommendations**   | Includes restaurants, attractions, and neighborhoods| Generic references (e.g., "local cafe", "bar")   |
| **Cultural Insight**        | Offers etiquette tips, safety advice, booking info | Minimal cultural context                         |
| **Repetition Handling**     | High variety across days                           | Some repetition due to short token constraint     |
| **Formatting**              | Professional-grade output with section headings    | Consistent format (Morning/Afternoon/Evening)    |
| **Speed on Apple M1**       | Fast (depends on API latency)                      | Fast on CPU or Metal (MPS) backend               |
| **Privacy & Cost**          | Data sent to external API; limited free usage      | Local processing; free and private               |
| **Customization**           | Limited to prompt variations                       | Fully controllable and modifiable                |
| **Best Use Case**           | Full-featured itinerary generator for end users    | Lightweight assistant or offline backup model    |

---

**Gemini** offers highly detailed and engaging itineraries enriched with **cultural context**, **transport suggestions**, and **polished formatting**—ideal for **production or consumer-facing applications**.

**TinyLLaMA**, while more basic and slightly repetitive due to its shorter token setting, delivers a **consistent, structured itinerary** suitable for **low-resource or privacy-sensitive scenarios**.

These two models highlight the trade-offs between **cloud-based intelligence** and **local autonomy**, illustrating how a **hybrid travel assistant system** can balance **depth with flexibility**.


# Web App Implementation 

Now that we've developed and tested our itinerary generation logic in the notebook, we’ve implemented the entire workflow as a user-friendly **Streamlit web application**. This web app brings together all the features and models from the notebook into a real-time, interactive interface.

The user flow is designed to be intuitive: the user first selects their **travel type** (e.g., solo, with family, or with friends), and based on this selection, the form dynamically reveals relevant group input fields such as the number of adults, children, or friends. Only then does the user complete the rest of the form to generate a personalized itinerary.

### Features Added to the Web App (`travel_assistant.py`)

This project extends the Jupyter-based assistant into a full-featured **Streamlit web application**, capable of real-time itinerary planning using both **cloud-based and offline LLMs**.

#### Core Features:
- **Gemini Integration (Online)**: Fast and well-structured itinerary generation via `gemini-1.5-pro-latest`.
- **TinyLLaMA Integration (Offline)**: Lightweight, privacy-focused LLM with local inference (runs entirely on-device).
- **Dynamic Travel Style Handling**: Users select a travel type first. If "with family" or "with friends" is selected, additional fields for number of adults, children, or friends appear immediately—*before* form submission.
- **Dietary Preference Support**: Includes a multi-select option for dietary needs (e.g., vegetarian, halal, gluten-free), which influences food-related suggestions in the itinerary.
- **Natural Dialogue Flow**: Captures all key user preferences: name, age, travel style, destination, duration, budget, interests, and dietary needs.
- **Prompt Builder**: Dynamically constructs a context-aware prompt tailored to the user’s full input and sends it to both LLMs.
- **City & Country Recognition**: Accepts both cities and countries as valid destinations using `geonamescache` and `pycountry`.
- **Destination Type Detection**: Informs the user whether the provided destination was identified as a city or a country.
- **Typo Handling for Destinations**: Automatically corrects common misspellings in destination names using `difflib`.
- **Translation Support (Gemini only)**: Users can instantly translate the itinerary into several languages (e.g., Spanish, French, Arabic).
- **Download Buttons**: Allows users to download the generated itinerary as a `.txt` file for offline access or sharing.
- **Model Speed Comparison**: Shows generation time for both Gemini and TinyLLaMA to help compare performance.
- **Offline Privacy Note**: Clearly communicates that TinyLLaMA runs locally and does not send any data to external servers.
- **Input Validation**: Ensures users fill out all required fields before allowing itinerary generation.
